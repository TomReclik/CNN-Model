# First Test to See Whether Images Are Read in Correctly

```python
%matplotlib inline
import os
import numpy as np
import matplotlib.pyplot as plt
import scipy.misc as misc
import xml.etree.ElementTree as ET
```

Load a single image and test if this is correctly loaded in

```python
PATH = "/home/tom/Data/LabeledDamages/CFK_def13_rep_1_2017-08-31/ETD__1_1-1.png"
img = misc.imread(PATH,flatten=True)
plt.imshow(img)
print img.shape
```

```python
PATH_XML = "/home/tom/Data/LabeledDamages/CFK_def13_rep_1_2017-08-31/ETD__1_1-1.xml"
tree = ET.parse(PATH_XML)
```

```python
root = tree.getroot()
print(root.tag)
```

```python
xmin = []
xmax = []
ymin = []
ymax = []

for ob in root:
    if ob.tag == "object":
        for bbox in ob:
            if bbox.tag == "bndbox":
                xmin.append(int(bbox[0].text))
                ymin.append(int(bbox[1].text))
                xmax.append(int(bbox[2].text))
                ymax.append(int(bbox[3].text))
                
print xmin,xmax,ymin,ymax
```

```python
damage = img[ymin[0]:ymax[0],xmin[0]:xmax[0]]
print damage.shape
plt.imshow(damage)
```

```python
for child in root.findall('object'):
    print child.tag, child.attrib
```
