# Find maximal bounding boxes

```python
PATH = "/home/tom/Data/LabeledDamages/CFK_def13_rep_1_2017-08-31/"

xmin = 1000000
ymin = 1000000
xmax = 0
ymax = 0
pxmax,pymax,pxmin,pymin = "","","",""
txmax,tymax,txmin,tymin = "","","",""

i=0
j=0

#IGNORE = ["ETD__8_9-1.xml", "ETD__13_5-1.xml", "ETD__11_14-1.xml"]

for PATH_XML in os.listdir(PATH):
    if PATH_XML.endswith(".xml") and not(PATH_XML in IGNORE):
        tree = ET.parse(PATH+PATH_XML)

        root = tree.getroot()

        for ob in root:
            if ob.tag == "object":
                for bbox in ob:
                    if bbox.tag == "name":
                        if bbox.text == "Artefact":
                            j += 1
                            break
                        else:
                            maxtype = bbox.text
                    if bbox.tag == "bndbox":
                        x = int(bbox[2].text)-int(bbox[0].text)
                        y = int(bbox[3].text)-int(bbox[1].text)
                        if x>1000:
                            i+=1
                        elif y>1000:
                            i+=1
                        else:
                            if x>xmax:
                                xmax = x
                                pxmax = PATH_XML
                                txmax = maxtype
                            if y>ymax:
                                ymax = y
                                pymax = PATH_XML
                                tymax = maxtype
                            if x<xmin:
                                xmin = x
                                pxmin = PATH_XML
                                txmin = maxtype                            
                            if y<ymin:
                                ymin = y
                                pymin = PATH_XML
                                tymin = maxtype

print "Maximal bounding box in x direction, xrange: ", xmax, " is of type: ", txmax
print "Maximal bounding box in y direction, yrange: ", ymax, " is of type: ", tymax
print "Minimal bounding box in x direction, xrange: ", xmin, " is of type: ", txmin
print "Minimal bounding box in y direction, yrange: ", ymin, " is of type: ", tymin
print "Path to xmax xml: ", pxmax
IGNORE.append(pxmax)
```
